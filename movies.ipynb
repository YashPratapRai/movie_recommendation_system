WordNetLemmatizer is a tool that uses WordNet to convert words into their base form (lemma).(running-->run)

```python
stop_words = stopwords.words('english')
print(stop_words)
len(stop_words)
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text) #If text contains anything other than letters or spaces, remove it completely.

  words = text.split() #To convert text (sentence) into a list of words(Computers can’t process sentences easily — they need tokens (words))
  #remove stopwords
  words = [word for word in words if word not in stop_words] #Keep only words that are NOT in stopwords list

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

✅ max_features=50000 → Dataset ke sab text me se top 50k most important words/phrases hi vocabulary me rakhta hai taaki model fast aur memory efficient rahe.

✅ ngram_range=(1,2) → Single words (unigrams) aur 2-word phrases (bigrams) dono ko features ke roop me include karta hai (jaise “science” + “science fiction”).

✅ TF-IDF → Text ko numerical vectors me convert karta hai based on word importance.

✅ TF (Term Frequency) → Word document me kitni baar aaya.

✅ IDF (Inverse Document Frequency) → Rare words ko zyada importance deta hai, common words ko kam.

✅ max_features → Vocabulary size limit karta hai (top important words hi rakhta hai).

✅ ngram_range → Single words aur phrases ko features me include karta hai.

### Cosine Similarity

✅ Cosine Similarity → Do vectors (documents/movies) ke beech similarity measure karta hai angle ke basis par.

✅ Range → Value −1 se 1 tak hoti hai, but text similarity me usually 0 to 1.

✅ 1 → Exactly same content
✅ 0 → No similarity

🧠 Formula
cosine similarity=A⋅B/|A||B|

👉 Dot product divided by magnitude

🎬 In recommendation systems

👉 Movie vectors (TF-IDF) ke beech similarity calculate karke similar movies recommend karta hai.

📊 Example intuition

Movie 1 → action adventure space
Movie 2 → action space

👉 High similarity

Movie 3 → romantic drama

👉 Low similarity
One line summary

👉 Cosine similarity vectors ke beech angle compare karke similarity score nikalta hai.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10):
    # Check if the movie title exists in our indices dictionary
    if title not in indices:
        return ['Movie not found']  # If not found, return message

    # Get the index of the given movie
    idx = indices[title]

    # Calculate cosine similarity between this movie vector
    # and all other movie vectors in the TF-IDF matrix
    # Result is a similarity score array
    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    # argsort() sorts indices based on similarity scores
    # [::-1] reverses to get highest similarity first
    # [1:n+1] skips the first one (itself) and takes top n similar movies
    similar_idx = sim_score.argsort()[::-1][1:n+1]

    # Return titles of the similar movies using their indices
    return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```
